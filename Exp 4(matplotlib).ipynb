```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [15, 30, 45, 10]
```

```python
# Line Plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, marker='o', linestyle='-')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(8, 4))
plt.bar(x, y, tick_label=labels)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(axis='y')
plt.show()
```
