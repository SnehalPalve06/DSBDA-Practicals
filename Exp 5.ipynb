```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd
```

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [15, 30, 45, 10]
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [10, 15, 20, 25, 30]
```

```python
# Line Plot
sns.set_style("whitegrid")
plt.figure(figsize=(8, 4))
sns.lineplot(x=x, y=y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(8, 4))
sns.barplot(x=labels, y=y)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(8, 4))
sns.scatterplot(x=scatter_x, y=scatter_y, color='blue')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
